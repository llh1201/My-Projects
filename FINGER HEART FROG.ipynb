```python
from PIL import Image 
from IPython.display import display 
import random
import json
import os
```

```python
background = ["Brown", "DeepBlue","Yellow","Green","Grey","Blue","Orange","Pink","Purple","Red","ShallowBlue"]
background_weights = [9,9,9,9,9,9,9,9,9,10,9]

body = ["bluebody", "greenbody", "orangebody", "pinkbody","purplebody","redbody","yellowbody"]
body_weights = [14,15,14,15,14,14,14]

eye = ["Sunglasses", "Small", "Big", "Point", "Happy"]
eye_weights = [70, 10, 5 , 1 , 14]

heart = ["purpleheart", "redheart", "orangeheart","pinkheart","yellowheart","greenheart","blueheart"]
heart_weights = [14,15,14,15,14,14,14]

mouth = ["short","smile","tongue","halfsmile"]
mouth_weights = [25,25,25,25]
```

```python
background_files = {
    "Brown": "bg1",
    "DeepBlue": "bg2",
    "Yellow": "bg3",
    "Green": "bg4",
    "Grey": "bg5",
    "Blue": "bg6",
    "Orange": "bg7",
    "Pink": "bg8",
    "Purple": "bg9",
    "Red": "bg10",
    "ShallowBlue": "bg11"
}

body_files = {
    "bluebody": "b1",
    "greenbody": "b2",
    "orangebody": "b3",
    "pinkbody": "b4",
    "purplebody": "b5",
    "redbody": "b6",
    "yellowbody": "b7"
}

eye_files = {
    "Sunglasses": "e1",
    "Small": "e2",
    "Big": "e3",
    "Point": "e4",
    "Happy": "e5"     
}

heart_files = {
    "purpleheart": "h1",
    "redheart": "h2",
    "orangeheart": "h3",
    "pinkheart": "h4",
    "yellowheart": "h5",
    "greenheart": "h6",
    "blueheart": "h7"
}


mouth_files = {
    "short": "m1",
    "smile": "m2",
    "tongue": "m3",
    "halfsmile": "m4",
}
```

```python
## Generate Traits

TOTAL_IMAGES = 5000 # Number of random unique images we want to generate

all_images = [] 

# A recursive function to generate unique image combinations
def create_new_image():
    
    new_image = {} #

    # For each trait category, select a random trait based on the weightings 
    new_image ["Background"] = random.choices(background, background_weights)[0]
    new_image ["Body"] = random.choices(body, body_weights)[0]
    new_image ["Eye"] = random.choices(eye, eye_weights)[0]
    new_image ["Heart"] = random.choices(heart, heart_weights)[0]
    new_image ["Mouth"] = random.choices(mouth, mouth_weights)[0]
    
    if new_image in all_images:
        return create_new_image()
    else:
        return new_image
    
    
# Generate the unique combinations based on trait weightings
for i in range(TOTAL_IMAGES): 
    
    new_trait_image = create_new_image()
    
    all_images.append(new_trait_image)
```

```python
# Returns true if all images are unique
def all_images_unique(all_images):
    seen = list()
    return not any(i in seen or seen.append(i) for i in all_images)

print("Are all images unique?", all_images_unique(all_images))
# Add token Id to each image
i = 0
for item in all_images:
    item["tokenId"] = i
    i = i + 1
   
print(all_images)
```

```python
background_count = {}
for item in background:
    background_count[item] = 0
    
body_count = {}
for item in body:
    body_count[item] = 0

eye_count = {}
for item in eye:
    eye_count[item] = 0
    
heart_count = {}
for item in heart:
    heart_count[item] = 0
    
mouth_count = {}
for item in mouth:
    mouth_count[item] = 0

for image in all_images:
    background_count[image["Background"]] += 1
    body_count[image["Body"]] += 1
    eye_count[image["Eye"]] += 1
    heart_count[image["Heart"]] += 1
    mouth_count[image["Mouth"]] += 1

print(background_count)
print(body_count)
print(eye_count)
print(heart_count)
print(mouth_count)
```

```python
os.mkdir(f'./images')

for item in all_images:

    im1 = Image.open(f'./Traits/background/{background_files[item["Background"]]}.png').convert('RGBA')
    im2 = Image.open(f'./Traits/body/{body_files[item["Body"]]}.png').convert('RGBA')
    im3 = Image.open(f'./Traits/eye/{eye_files[item["Eye"]]}.png').convert('RGBA') 
    im4 = Image.open(f'./Traits/heart/{heart_files[item["Heart"]]}.png').convert('RGBA')
    im5 = Image.open(f'./Traits/mouth/{mouth_files[item["Mouth"]]}.png').convert('RGBA')

    #Create each composite
    com1 = Image.alpha_composite(im1, im2)
    com2 = Image.alpha_composite(com1, im3)
    com3 = Image.alpha_composite(com2, im4)
    com4 = Image.alpha_composite(com3, im5)                

    #Convert to RGB
    rgb_im = com4.convert('RGB')
    file_name = str(item["tokenId"]) + ".png"
    rgb_im.save("./images/" + file_name)
```
